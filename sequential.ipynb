```python
q1 = df['horsepower'].quantile(q=0.25)
q3 = df['horsepower'].quantile(q=0.75)  
iqr = q3 - q1   
lower = (q1 - 1.5 * iqr) 
upper = (q3 + 1.5 * iqr)
```

```python
df['horsepower'].clip(lower, upper)
```
